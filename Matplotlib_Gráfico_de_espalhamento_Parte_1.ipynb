```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,5,11)
```

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
#gráfico de y em função de x
plt.scatter(X,Y)
plt.show()
```

```python
#inserindo nome nos eixos
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
#inserindo titulo no gráfico
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
#dois gráficos na mesma imagem
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
#alterando cores
plt.scatter(X,Y,color='black')
plt.scatter(X,Z,color='red')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
#inserindo legenda
plt.scatter(X,Y,color='black',label='Y(X)')
plt.scatter(X,Z,color='red',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
