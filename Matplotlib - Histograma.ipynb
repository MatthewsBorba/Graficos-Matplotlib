```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Criando variável X com números aleatórios
X = np.random.rand(2000)
```

```python
#Criando histograma
plt.hist(X)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.show()
```

```python
#Alterando número de bins
plt.hist(X,bins=15)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=15')
plt.show()
```

```python
plt.hist(X,bins=50)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
#Alterando cores
plt.hist(X,bins=50,color='purple')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
#Multiplos histogramas na mesma imagem
X1 = np.random.normal(0,1,size = 800)
X2 = np.random.normal(3,1,size = 800)
X3 = np.random.normal(5,1,size = 800)
```

```python
plt.hist(X1,bins=10,color='red')
plt.hist(X2,bins=10,color='blue')
plt.hist(X3,bins=10,color='green')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Multiplos histogramas')
plt.show()
```

```python
#Distribuição cumulativa
X = np.random.rand(2000)
```

```python
plt.hist(X,bins=20,color='red', edgecolor='blue',cumulative=True)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```
