```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável X
X = np.linspace(0,5,11)
```

```python
# Criando função Y(X)
Y = np.power(X-2,2.0)
```

```python
#Criando função Z(X)
Z = np.exp(X/2.0)
```

```python
# Criando grid do Y
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle = '-' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle = '--' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```

```python
# Criando grid do Z
plt.plot(X,Y,label='$Z(X)=(X/2)$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Z')
plt.title('Z(X)')
plt.grid(linestyle = '-' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```
