```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,5,11)
```

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
#alterando o tamanho do ponto
plt.scatter(X,Y,s=10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
#dois gráficos com pontos de tamanhos diferentes
plt.scatter(X,Y,s=10,label='Y(X)')
plt.scatter(X,Z,s=40,label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

```python
#alkterando a opacidade do ponto
plt.scatter(X,Y,s=10,label='Y(X)',alpha=0.8)
plt.scatter(X,Z,s=40,label='Z(X)',alpha=0.4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```
