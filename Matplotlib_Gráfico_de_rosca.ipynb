```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando labels
labels = ["Grupo 1","Grupo 2","Grupo 3","Grupo 4","Grupo 5"]
```

```python
#Criando lista de valores
valores = [15,40,37,50,21]
```

```python
#Criando lista com cores
cores = ['red','green','blue','orange','cyan']
```

```python
#definindo raio
raio = 0.7
```

```python
#criando gráfico de rosca
plt.pie(valores,labels=labels,colors=cores)
circulo = plt.Circle((0,0),raio,color='white')
p = plt.gcf()
p.gca().add_artist(circulo)
plt.show()
```

```python
#adicionando porcentagem
plt.pie(valores,labels=labels,colors=cores,autopct='%1.1f%%')
circulo = plt.Circle((0,0),raio,color='white')
p = plt.gcf()
p.gca().add_artist(circulo)
plt.show()
```
